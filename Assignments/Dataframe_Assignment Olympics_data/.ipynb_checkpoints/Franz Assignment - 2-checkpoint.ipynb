```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### Read data from olympics.csv

```python
olympics = pd.read_csv('olympics.csv', index_col=0, skiprows=1)
```

```python
olympics
```

##### Display list of Country names which has never won a Gold in Summer

```python
nogold_summer = olympics[olympics["01 !"] == 0]
pd.DataFrame(nogold_summer.index,columns =["No Gold (Summer)"])
```

##### Display list of Country names which has never won a Gold in either Summer or Winter Games

```python
nogold_summer_winter = olympics[(olympics["01 !"] == 0) | (olympics["01 !.1"] == 0)] 
pd.DataFrame(nogold_summer_winter.index, columns =["No Gold (Summer or Winter)"])
```

##### Display name of the Country which has never won a Gold in Summer Games but won in Winter Games

```python
nogold_summer_gold_winter = olympics[(olympics["01 !"] == 0) & ~(olympics["01 !.1"] == 0)] 
pd.DataFrame(nogold_summer_gold_winter.index, columns =["No Gold (Summer) - Gold (Winter)"])
```

##### Which country has won maximum number of medals

```python
olympics2 = pd.read_csv('olympics.csv', index_col=0, skiprows=[0,148])
olympics2[olympics2["Combined total"]==olympics2["Combined total"].max()]

```

##### Which country has won the most gold medals in winter games? Plot a pie chart showing Gold, Silver, Bronze totals for this country

```python
olymp = olympics2[olympics2["Total.1"]==olympics2["Total.1"].max()]

gold = olymp["01 !.1"]
silver = olymp["02 !.1"]
bronze = olymp["03 !.1"]

fig = plt.figure(figsize = (10,7))
obj = plt.pie([gold,silver,bronze], autopct = "%1.1f%%", shadow = True,labels = ["gold","silver","bronze"], colors = ["gold","silver","brown"])
plt.title("Medal Tally", pad = 15, fontdict = {"fontsize":25, "fontweight":2, "color":"Blue"})
```

##### Which country had the biggest difference between their summer and winter gold medal counts?

```python
olympics2["Total Summer and Winter Gold"] = np.where(olympics2["01 !"] > olympics2['01 !.1'], olympics2["01 !"] - olympics2["01 !.1"], olympics2["01 !.1"] - olympics2["01 !"])
onegold_summer_winter = olympics2[(olympics2["01 !"] > 0) & (olympics2["01 !.1"] > 0)]
olympics2[olympics2["Total Summer and Winter Gold"]==olympics2["Total Summer and Winter Gold"].max()]                               
```

##### Which country has the biggest difference between their summer gold medal counts and winter gold medal counts relative to their total gold medal count?
 
Only include countries that have won at least 1 gold in both summer and winter.

```python
olympics2["Total Summer and Winter Gold"] = np.where(olympics2["01 !"] > olympics2['01 !.1'], olympics2["01 !"] - olympics2["01 !.1"], olympics2["01 !.1"] - olympics2["01 !"])
onegold_summer_winter = olympics2[(olympics2["01 !"] > 0) & (olympics2["01 !.1"] > 0)]
onegold_summer_winter["Relative Diff"] = onegold_summer_winter["01 !.2"] - onegold_summer_winter["Total Summer and Winter Gold"]
onegold_summer_winter[onegold_summer_winter["Relative Diff"]==onegold_summer_winter["Relative Diff"].max()]  

```

##### Write a function that creates a Series called "Points" which is a weighted value where each gold medal counts for 3 points, silver medals for 2 points, and bronze medals for 1 point. The function should return only the column (a Series object) which you created, with the country names as indices.

```python
def pointsfunc(x):
    x.Points = (x["01 !.2"]*3) + (x["02 !.2"]*2) + (x["03 !.2"]*1)
    return x.Points   

olympics2 = pd.read_csv('olympics.csv', index_col=0, skiprows=[0,148])
olympics2["Points"]=pd.Series()

pointsfunc(olympics2)
```

##### Identify top 5 medal winners (countries). Store their details in a new Data frame. Plot a bar chart showing their Gold, Silver, Bronze counts



```python
olympics2 = pd.read_csv('olympics.csv', index_col=0, skiprows=[0,148])
olympics2["Combined total"].sort_values

olympics2.reset_index()
olympics_top5 = olympics2.sort_values(by='Combined total', ascending=False)
top5 = olympics_top5[0:5]

plot = top5[["01 !.2","02 !.2","03 !.2"]].plot(kind="bar", figsize = (20,5), rot = 30, color = ["Gold","Silver","Brown"], title = "Gold, Silver, Bronze counts", fontsize = 15, edgecolor = "Blue", alpha = 0.6)
```
